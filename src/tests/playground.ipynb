```python
import requests
import pandas as pd

def search_google(query, api_key, cx):
    # Define the endpoint
    url = "https://www.googleapis.com/customsearch/v1"

    # Define the parameters
    params = {
        "key": api_key,
        "cx": cx,
        "q": query,
    }

    # Send the GET request
    response = requests.get(url, params=params)

    # Convert the response to JSON
    data = response.json()

    return data

def parse_results(data):
    results = []

    # Loop through the items in the data
    for item in data["items"]:
        # Extract the title, link, and snippet
        title = item["title"]
        link = item["link"]
        snippet = item["snippet"]

        # Append the result to the results list
        results.append({
            "Title": title,
            "Link": link,
            "Snippet": snippet,
        })

    # Convert the results to a DataFrame
    df = pd.DataFrame(results)

    return df

# Define your API key and Programmable Search Engine ID
api_key = "AIzaSyC49nX1Wzjqvwb2gWnXMcFj-pjNzNRSyNE"
cx = "434a55df8dcf54c1e"

# Perform the search
data = search_google("Plant based protein tradeshows", api_key, cx)

# Parse the results
df = parse_results(data)

# Print the DataFrame
print(df)
```
